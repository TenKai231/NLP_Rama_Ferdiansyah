```python
from google_play_scraper import reviews, Sort
import pandas as pd
```

```python
result, continuation_token = reviews(
    'com.shopee.id',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=5000
)
```

```python
df = pd.DataFrame(result)
```

```python
df = df[['content', 'score']]
```

```python
def labeling(score):
    if score <= 2:
        return 'negatif'
    elif score == 3:
        return 'netral'
    else:
        return 'positif'

df['label'] = df['score'].apply(labeling)
```

```python
df.to_csv('review_shopee.csv', index=False)
```
