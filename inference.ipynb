```python
import pickle
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model
```

```python
with open('model/tfidf.pkl', 'rb') as f:
    tfidf = pickle.load(f)
```

```python
new_text = [
    "aplikasinya sangat bagus dan cepat"
]
```

```python
new_text_tfidf = tfidf.transform(new_text)
```

```python
model = load_model(
    'model/sentiment_model.h5',
    compile=False
)
```

```python
# Predict
prediction = model.predict(new_text_tfidf)
```

```python
prediction = model.predict(new_text_tfidf)

predicted_class = prediction.argmax(axis=1)

label_map = {
    0: "negatif",
    1: "netral",
    2: "positif"
}

print("Prediction :", label_map[predicted_class[0]])
```

```python
confidence = prediction.max()

print("Confidence :", confidence)
```
